```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.1
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        md = -(2/len(x)) * sum(x * (y - y_predicted))
        bd = -(2/len(x)) * sum(y - y_predicted)
        cost = (1/n) * sum((y - y_predicted) ** 2)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        # Print every 100 iterations to avoid too much output
        if i % 100 == 0:
            print(f"m: {m_curr:.4f}, b: {b_curr:.4f}, iteration: {i}, cost: {cost:.4f}")
    
    return m_curr, b_curr

# Define data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Call function
m_final, b_final = gradient_descent(x, y)
print(f"Final: m={m_final:.4f}, b={b_final:.4f}")
```
